```python
import matplotlib.pyplot as plt 
import numpy as np
import pandas as pd
%matplotlib inline
```

```python
#Goal: Determine if a home advantage exists in soccer
soccer = pd.read_csv('SoccerLEagues.csv')
plt.figure(figsize=(8,5))

plt.title('Home vs Away Game Wins across Multiple Soccer Leagues', fontdict={'fontweight':'bold', 'fontsize': 10})

#Calclulate relation between Home winning ratio 
#and away winning ratio across all teams 
#Clean the data because some is unusable
i = 0
print("Starting Cleaning Process...")
while i < int(soccer.shape[0]):
	if (soccer.AwayWins[i] > soccer.Games[i]) or (soccer.HomeWins[i] > soccer.Games[i]):
		soccer.drop(index=i, axis=0, inplace = True)	
	i = i + 1
print("Complet")
HPCT = soccer.HomeWins/soccer.Games
print(HPCT.head())
APCT = soccer.AwayWins/soccer.Games
mean_HPCT = HPCT.mean()
mean_APCT = APCT.mean()


plt.scatter(HPCT, APCT, s = 1, color = "blue", label = "Data points")
plt.scatter(mean_HPCT, mean_APCT, s = 30, color = 'red', label = "Mean Value")
x = [0,1]
y = [0,1]
plt.plot(x, y, color = 'purple')
plt.xlabel("Normalied Home Wins")
plt.ylabel("Normalized Away Wins")
plt.xlim([0,0.6])
plt.ylim([0,0.6])
plt.legend()

```
